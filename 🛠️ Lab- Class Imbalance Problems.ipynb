## Define the predictor and target variables

Define `X` and `y` and perform a standard train-test split. Assign 25% to the test set and `random_state` to 0. 

```python
# Your code here
# Import necessary packages
from sklearn.model_selection import train_test_split

# Define the target variable y and the feature matrix X
y = df['Class']
X = df.drop('Class', axis=1)

# Perform a standard train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Display the shapes of the resulting datasets
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

Find the class imbalance in the training and test sets: 

```python
# Find the class imbalance in the training set
print("Training set class distribution:")
print(y_train.value_counts())
print("\nTraining set class distribution (percentage):")
print(y_train.value_counts(normalize=True) * 100)

print('\n')

# Find the class imbalance in the test set
print("Test set class distribution:")
print(y_test.value_counts())
print("\nTest set class distribution (percentage):")
print(y_test.value_counts(normalize=True) * 100)
```

## Create an initial model

As a baseline, train a vanilla logistic regression model. Then plot the ROC curve and print out the AUC. We'll use this as a comparison for how our future models perform.

```python
# Initial Model
# Import necessary packages
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Initial Model
logreg = LogisticRegression(solver='liblinear')

# Fit the model to the training data
logreg.fit(X_train, y_train)

# Probability scores for the test set
y_score = logreg.predict_proba(X_test)[:, 1]

# False positive rate and true positive rate
fpr, tpr, thresholds = roc_curve(y_test, y_score)

# Compute the AUC
roc_auc = auc(fpr, tpr)

# Seaborn's beautiful styling
sns.set_style('darkgrid', {'axes.facecolor': '0.9'})

# Print AUC
print(f'Initial Model AUC: {roc_auc:.2f}')

# Plot the ROC curve
plt.figure(figsize=(10, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()

```

## Tune the model 

Try some of the various techniques proposed to tune your model. Compare your models using AUC and ROC curve.

```python
# Import necessary packages
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('creditcard.csv.gz', compression='gzip')

# Define the target variable y and the feature matrix X
y = df['Class']
X = df.drop('Class', axis=1)

# Perform a standard train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Define the range of C values
C_param_range = [0.001, 0.01, 0.1, 1, 10, 100]
names = [0.001, 0.01, 0.1, 1, 10, 100]
colors = sns.color_palette('Set2')

plt.figure(figsize=(10, 8))

for n, c in enumerate(C_param_range):
    # Fit a logistic regression model with the current C value
    logreg = LogisticRegression(C=c, solver='liblinear')
    model_log = logreg.fit(X_train, y_train)
    print(model_log)  # Preview model params

    # Predict probabilities for the test set
    y_score = logreg.predict_proba(X_test)[:, 1]

    # Compute the ROC curve and AUC
    fpr, tpr, thresholds = roc_curve(y_test, y_score)
    roc_auc = auc(fpr, tpr)
    
    print('AUC for {}: {}'.format(names[n], roc_auc))
    print('-------------------------------------------------------')
    
    # Plot the ROC curve
    lw = 2
    plt.plot(fpr, tpr, color=colors[n], lw=lw, label='ROC curve Normalization Weight: {}'.format(names[n]))

# Plot the diagonal line for reference
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.yticks([i/20.0 for i in range(21)])
plt.xticks([i/20.0 for i in range(21)])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```
